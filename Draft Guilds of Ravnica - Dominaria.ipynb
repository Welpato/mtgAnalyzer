```python
import src.mtgBaseClass as mbc
```

In this analisys I want compare and get usefull information for drafting in the following sets: Guilds of Ravnica and Dominaria.
    

```python
setList = ['Guilds of Ravnica', 'Dominaria']
compare = mbc.mtgBase(setList)
```

# Total cards per Rarity

```python
#Plot the total of cards of each set
compare.plotCardsTotal(compare.cardSetList)
```

# Median CMC per rarity

```python
#Plot the median CMC per rarity of each set
compare.plotMedianCmc(compare.cardSetList)
```

# Total of Creatures

```python
setList = compare.filterCardSetListBy(compare.cardSetList, compare.card.type, 'Creature')
compare.plotCardsTotal(setList)
```

# White Cards with double cost

```python
colorRepeat = [compare.colorPie.white.colorShort]
setList = compare.filterCardSetListBy(compare.cardSetList,compare.card.mana_cost,compare.regex.regexValueRepeat(colorRepeat, 2,"\{", "\}"))
```

```python
compare.plotCardsTotal(setList)
```

```python
compare.plotCardSetList(setList)
```
