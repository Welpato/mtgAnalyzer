You can filter the cardSetList using the function "filterCardSetListBy" and then selecting the column that you want to filter
and then the value that will be filtred.

In this case it will filter by all the cards that have the word "soldiers" in the flavor text


```python
soldiersFiltering = compare.filterCardSetListBy(compare.cardSetList,compare.card.flavor,'soldiers')
compare.plotCardSetList(soldiersFiltering)
```

You can also do more complex filtering using function like 'returnColorIdentity'

```python
colorIdentity = [compare.colorPie.white.colorShort, compare.colorPie.red.colorShort]
setList = compare.filterCardSetListBy(compare.cardSetList,compare.card.color_identity,compare.regex.regexValueExists(colorIdentity))
compare.plotCardSetList(setList)
```

Is also possible to apply a filter over a filter.

In this case is filtering the result of the cell above by cards with the word 'Izzet' in the name

```python
setList = compare.filterCardSetListBy(setList, compare.card.name,'Izzet')
compare.plotCardSetList(setList)
```

You can filter for color repetition on the mana cost, with a minimal repetition.

Like in this case it plots all the cards that have the at least 2 black mana in the cost

```python
colorRepeat = [compare.colorPie.black.colorShort]
setList = compare.filterCardSetListBy(compare.cardSetList,compare.card.mana_cost,compare.regex.regexValueRepeat(colorRepeat, 2,"\{", "\}"))
compare.plotCardSetList(setList)
```
